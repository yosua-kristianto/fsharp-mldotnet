# Daily Show Guests Cleaned

Data ini berisi data figur yang diundang dalam sebuah acara (show) sebagai tamu.

```python
import pandas
import matplotlib.pyplot as plt
```

```python
dataframe = pandas.read_csv("data/daily_show_guests_cleaned.csv");

# Data exploration
dataframe.head()
```

# Data Pre-Processing

Dalam eksplorasi singkat pada data daily_show_guests_cleaned, didapati bahwa datanya tidak clean. Beberapa data kosong, dan ada kolom yang perlu di fix.

> 1. Untuk data yang tidak clean, dalam hal ini berarti memiliki nilai kosong atau Not Available, maka treatment yang dilakukan adalah data tersebut di drop
> 2. Ditemukan bahwa isian data Raw_Guest_List dan GoogleKnowledge_Occupation tidak konsisten. Maka seluruh nilai pada kolom tersebut akan diterapkan UPPERCASE.

```python
# Untuk setiap data yang Not Available, Sikat abis.
dataframe = dataframe.dropna()

# GoogleKnowledge_Occupation force uppercase
dataframe["GoogleKnowlege_Occupation"] = dataframe["GoogleKnowlege_Occupation"].str.upper()
dataframe["Raw_Guest_List"] = dataframe["Raw_Guest_List"].str.upper()

dataframe.head()
```

```python
# Get all years unique
years = dataframe["YEAR"]
years = years.unique()

x = years
y = []

for i in x:
    y.append(dataframe.loc[dataframe["YEAR"] == i].shape[0])
    i = str(i)

plt.bar(x, y)

plt.title("Visualisasi jumlah show tahun 2000 hingga 2015")
plt.xlabel("Tahun")
plt.ylabel("Total show")

plt.show()
```

```python
# Get all years unique
years = dataframe["Group"]
years = years.unique()

x = years
y = []

for i in x:
    y.append(dataframe.loc[dataframe["Group"] == i].shape[0])
    i = str(i)

plt.bar(x, y)

plt.title("Visualisasi jumlah jenis show guest diundang sejak 2000 - 2015")
plt.xlabel("Group")
plt.ylabel("Jumlah")

plt.xticks(x, rotation = "vertical")

plt.show()
```
