# Insurance Premium

Data ini memberikan beberapa insight:

1. Jumlah kecelakaan mobil (spesifiknya, car crash akibat tabrakan mobil)
2. Jumlah biaya yang dikeluarkan asuransi untuk me-cover kerusakan kendaraan

Berdasarkan riset yang dilakukan pada data insurance_premiums, didapati bahwa:

| No | Kolom | Deskripsi |
|---|---|---|
| 1 | State | Negara Bagian pencatatan data |
| 2 | fatal_collisions | Angka rata-rata tabrakan fatal pada negara bagian, dalam satuan tertentu. Tidak diketahui pasti satuan apa yang digunakan. Untuk menghindari kesalahan dalam pengolahan visualisasi data, kolom ini akan di drop. |
| 3 | fatal_collisions_speeding | Jumlah tabrakan fatal akibat ngebut. |
| 4 | fatal_collisions_alc | Jumlah tabrakan fatal akibat konsumsi alkohol (Driving Under Influences). |
| 5 | fatal_collisions_not_distracted | Jumlah tabrakan fatal yang terjadi akibat kelalaian pengemudi. |
| 6 | fatal_collisions_no_hist | Jumlah tabrakan fatall yang terjadi, namun pengemudi tidak pernah mengalami kecelakaan sebelumnya. |
| 7 | premiums | Jumlah rata-rata premi asuransi kendaraan yang dibayarkan pemilik kendaraan. |
| 8 | insurance_losses | Jumlah rata-rata coverage yang dibayarkan perusahaan asuransi kepada pemilik kendaraan. |
| 9 | Region | Regional State. Contoh, California adalah Barat, Florida adalah Selatan. | 

```python
import pandas
import matplotlib.pyplot as plt
```

```python
dataframe = pandas.read_csv("data/insurance_premiums.csv")

# Data exploration
dataframe.head()
```

# Data Pre-Processing

Seperti yang disebutkan diawal, data fatal_collisions akan di drop

```python
dataframe = dataframe.drop(columns = ["fatal_collisions"])
```

```python
# Visualisasi jumlah angka kecelakaan akibat ngebut disetiap state

x = dataframe["State"]
y = dataframe["fatal_collisions_speeding"]

plt.figure(figsize=(15, 15))
plt.bar(x, y)
plt.title("Visualisasi jumlah angka kecelakaan akibat ngebut disetiap state")
plt.xlabel("State")
plt.ylabel("Data Kecelakaan")
plt.xticks(x, rotation = "vertical")

for i, j in zip(x, y):
    plt.text(i, j, str(j))

plt.show()
```

```python
# Visualisasi jumlah angka kecelakaan akibat DUI disetiap state

x = dataframe["State"]
y = dataframe["fatal_collisions_alc"]

plt.figure(figsize=(15, 15))
plt.bar(x, y)
plt.title("Visualisasi jumlah angka kecelakaan akibat DUI disetiap state")
plt.xlabel("State")
plt.ylabel("Data Kecelakaan")
plt.xticks(x, rotation = "vertical")

for i, j in zip(x, y):
    plt.text(i, j, str(j))

plt.show()
```

```python
# Visualisasi jumlah angka kecelakaan akibat DUI disetiap state

x = dataframe["State"]
y = dataframe["fatal_collisions_not_distracted"]

plt.figure(figsize=(15, 15))
plt.bar(x, y)
plt.title("Visualisasi jumlah angka kecelakaan akibat kelalaian pengemudi disetiap state")
plt.xlabel("State")
plt.ylabel("Data Kecelakaan")
plt.xticks(x, rotation = "vertical")

for i, j in zip(x, y):
    plt.text(i, j, str(j))

plt.show()
```

```python
# Visualisasi jumlah angka kecelakaan akibat DUI disetiap state

x = dataframe["State"]
y = dataframe["fatal_collisions_no_hist"]

plt.figure(figsize=(15, 15))
plt.bar(x, y)
plt.title("Visualisasi jumlah tabrakan fatal yang terjadi, namun pengemudi tidak pernah mengalami kecelakaan sebelumnya.")
plt.xlabel("State")
plt.ylabel("Data Kecelakaan")
plt.xticks(x, rotation = "vertical")

for i, j in zip(x, y):
    plt.text(i, j, str(j))

plt.show()
```

```python
# Visualisasi angka rata-rata kerugian perusahaan asuransi dalam me-cover kerugian akibat kecelakaan

x = dataframe["State"]
y = dataframe["insurance_losses"]

plt.figure(figsize=(15, 15))
plt.bar(x, y)
plt.title("Visualisasi angka rata-rata kerugian perusahaan asuransi dalam me-cover kerugian akibat kecelakaan")
plt.xlabel("State")
plt.ylabel("Data Kecelakaan")
plt.xticks(x, rotation = "vertical")

for i, j in zip(x, y):
    plt.text(i, j, str(j))

plt.show()
```
